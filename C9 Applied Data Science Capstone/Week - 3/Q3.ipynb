```python
import pandas as pd
import folium
```

```python
torontoData = pd.read_csv('List_of_postal_codes_of_Canada_with_latlong.csv', index_col = 0)

torontoData.head()
```

```python
torontoData.shape
```

```python
x = torontoData['Latitude']
y = torontoData['Longitude']

inpData = torontoData[['Latitude', 'Longitude']]
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=4, random_state = 7)
```

```python
model.fit(inpData)
```

```python
torontoData['Cluster'] = model.labels_
```

```python
torontoData['Cluster'].value_counts()
```

```python
map_out = folium.Map(location=[43.65, -79.4], zoom_start=12)

colors = ['red', 'green', 'blue', 'yellow']

for lat, lon, bor, cluster in zip(torontoData['Latitude'], torontoData['Longitude'], torontoData['Borough'], torontoData['Cluster']):
    label = folium.Popup(bor, parse_html=True)
    folium.CircleMarker(
        [lat, lon], radius=5,
        popup=label, color='black',
        fill=True, fill_color = colors[cluster],
        fill_opacity=0.7).add_to(mapout)  

mapout
```

```python
mapout.save('finalOutput.png')
```
